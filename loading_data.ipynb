Pretaitement du dataset

```python
# Définir les types de carburants à garder
fuel_allowed = ["essence", "gasoil", "hybride", "e85"]

train_df = train_df[train_df["fuel_type"].isin(fuel_allowed)]

#Encodage des carburants
fuel_mapping = {"essence": 0, "gasoil": 1, "hybride": 2, "e85": 3}
train_df["fuel_type"] = train_df["fuel_type"].map(fuel_mapping)

label_encoder_brand = LabelEncoder()
label_encoder_model = LabelEncoder()

train_df["brand"] = label_encoder_brand.fit_transform(train_df["brand"])
train_df["model"] = label_encoder_model.fit_transform(train_df["model"])


# Fonction pour extraire uniquement le nombre de chevaux
def extract_hp(engine_str):
    match = re.search(r"(\d+\.?\d*)HP", str(engine_str))
    print(match.group(1))
    return float(match.group(1)) if match else None 



train_df["engine_hp"] = train_df["engine"].apply(extract_hp)

train_df.drop(columns=["engine"], inplace=True)

#Supprimer les lignes avec des informations manquantes
# train_df.dropna(inplace=True)


train_df.to_csv("data/train_cleaned.csv", index=False)
```
