Pretaitement du dataset

```python
# Définir les types de carburants à garder
print(len(train_df))
fuel_allowed = ["Gasoline", "Diesel", "Hybrid", "E85 Flex Fuel"]
train_df = train_df[train_df["fuel_type"].isin(fuel_allowed)]


#Encodage des carburants
fuel_mapping = {"Gasoline": 0, "Diesel": 1, "Hybrid": 2, "E85 Flex Fuel": 3}
train_df["fuel_type"] = train_df["fuel_type"].map(fuel_mapping)



label_encoder_brand = LabelEncoder()
label_encoder_model = LabelEncoder()

train_df["brand"] = label_encoder_brand.fit_transform(train_df["brand"])
train_df["model"] = label_encoder_model.fit_transform(train_df["model"])


# Fonction pour extraire uniquement le nombre de chevaux
def extract_hp(engine_str):
    match = re.search(r"(\d+\.?\d*)HP", str(engine_str))
    return float(match.group(1)) if match else None 

train_df["engine_hp"] = train_df["engine"].apply(extract_hp)
train_df.drop(columns=["engine"], inplace=True)

#Garder l'age de la voiture plutot que son année afin de réduire la tailles des entrées
train_df["age"] = 2025 - train_df["model_year"]
train_df.drop(columns=["model_year"], inplace=True)
train_df.drop(columns=["transmission"], inplace=True)
train_df.drop(columns=["ext_col"], inplace=True)
train_df.drop(columns=["int_col"], inplace=True)
train_df.drop(columns=["accident"], inplace=True)
train_df.drop(columns=["clean_title"], inplace=True)
train_df.drop(columns=["id"], inplace=True)

#Supprimer les lignes avec des informations manquantes
train_df.dropna(inplace=True)
print(len(train_df))

def remove_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR  
    out_data = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    df = df.drop(out_data.index).reset_index(drop=True)
    return df


train_df = remove_outliers(train_df,"milage")
train_df = remove_outliers(train_df,"price")


train_df.to_csv("data/train_cleaned.csv", index=False)
```

```python
# Définir les types de carburants à garder
print(len(test_df))
fuel_allowed = ["Gasoline", "Diesel", "Hybrid", "E85 Flex Fuel"]
test_df = test_df[test_df["fuel_type"].isin(fuel_allowed)]


#Encodage des carburants
fuel_mapping = {"Gasoline": 0, "Diesel": 1, "Hybrid": 2, "E85 Flex Fuel": 3}
test_df["fuel_type"] = test_df["fuel_type"].map(fuel_mapping)

label_encoder_brand = LabelEncoder()
label_encoder_model = LabelEncoder()

test_df["brand"] = label_encoder_brand.fit_transform(test_df["brand"])
test_df["model"] = label_encoder_model.fit_transform(test_df["model"])


# Fonction pour extraire uniquement le nombre de chevaux
def extract_hp(engine_str):
    match = re.search(r"(\d+\.?\d*)HP", str(engine_str))
    return float(match.group(1)) if match else None 

test_df["engine_hp"] = test_df["engine"].apply(extract_hp)
test_df.drop(columns=["engine"], inplace=True)

test_df["age"] = 2025 - test_df["model_year"]
test_df.drop(columns=["model_year"], inplace=True)

test_df.drop(columns=["transmission"], inplace=True)
test_df.drop(columns=["ext_col"], inplace=True)
test_df.drop(columns=["int_col"], inplace=True)
test_df.drop(columns=["accident"], inplace=True)
test_df.drop(columns=["clean_title"], inplace=True)
test_df.drop(columns=["id"], inplace=True)

#Supprimer les lignes avec des informations manquantes
test_df.dropna(inplace=True)
print(len(test_df))


def remove_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR  
    out_data = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    df = df.drop(out_data.index).reset_index(drop=True)
    return df


test_df = remove_outliers(test_df,"milage")
test_df = remove_outliers(test_df,"price")


test_df.to_csv("data/test_cleaned.csv", index=False)
```
